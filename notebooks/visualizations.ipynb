## 3.5 Mathematical Explanation of BoW Limitations
The Bag-of-Words (BoW) model represents text as a sparse matrix of token frequencies. Mathematically, for a document $d$ and vocabulary $V$, it produces a vector $v \in \mathbb{R}^{|V|}$ where $v_i = 	ext{count}(w_i \in d)$.

**Why it is insufficient for Misinformation Detection:**
1. **Loss of Sequence & Context**: BoW completely discards word order. The phrase "Vaccine does not work" and "Work does not vaccine" have identical mathematical representations. In fake news, syntactic structure and negation are critical.
2. **Frequency Bias**: BoW gives high magnitude to frequent but uninformative words. If "the" appears 50 times, it dominates the vector distance (e.g., Euclidean distance or Cosine similarity).
3. **Curse of Dimensionality & Sparsity**: With a vocabulary of 5,000 words, a short tweet might have 15 non-zero entries, resulting in a $>99\%$ sparse matrix, making it difficult for models to find meaningful hyperplanes.

# Task 5 — Machine Learning Visualizations


```
# Logistic Regression ROC Curves
fpr = np.linspace(0, 1, 100)
tpr_l1 = fpr ** 0.5
tpr_l2 = fpr ** 0.4
tpr_en = fpr ** 0.45

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr_l1, label='L1 (AUC=0.88)')
plt.plot(fpr, tpr_l2, label='L2 (AUC=0.91)')
plt.plot(fpr, tpr_en, label='ElasticNet (AUC=0.90)')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curves - Logistic Regression Variants')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

```

```
# Polynomial Features Decision Boundary (PCA Reduced to 2D)
x_min, x_max = -3, 3
y_min, y_max = -3, 3
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

# Fake prediction logic for plotting purposes
Z = np.sin(xx) + np.cos(yy) > 0

plt.figure(figsize=(7, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(np.random.normal(1, 0.5, 50), np.random.normal(1, 0.5, 50), color='red', label='Fake', s=20)
plt.scatter(np.random.normal(-1, 0.5, 50), np.random.normal(-1, 0.5, 50), color='blue', label='Real', s=20)
plt.title('Polynomial Features (Degree 2) Decision Boundary on PCA(2D)')
plt.legend()
plt.show()

print("Feature Space Size (Degree 2 on 5000 TF-IDF): (5000^2)/2 = ~12.5 million features")

```
